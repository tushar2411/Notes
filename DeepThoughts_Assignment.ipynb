## Week 1 


```python
parameter =['Q1','Q2','Q3','Q4','Q5']
neha=[10,9,9,9,9]
amit =[8,7,8,8,8]
sachin=[6,4,7,7,6]

barWidth = 0.25

r1=range(len(parameter))
r2=[x+barWidth for x in r1]
r3=[x+barWidth for x in r2]
# colors = ['#FFC300', '#FF5733', '#C70039']
fig, ax = plt.subplots(figsize=(15, 8))


plt.bar(r1, neha, color='#FFC300', width=barWidth, edgecolor='white', label='Neha')
plt.bar(r2, amit, color='#FF5733', width=barWidth, edgecolor='white', label='Amit')
plt.bar(r3, sachin, color='#C70039', width=barWidth, edgecolor='white', label='Sachin')

plt.xlabel('Parameters')
plt.ylabel('Scale 1 - 10')
plt.grid(True)
plt.legend()
plt.title('Efforts in Completing Projects and Personal Growth')

plt.show()
```
